what is a classifatcion problem :
a classification problem is a type of supervised learning problme in machine learning wehre the goal is to categorize data points into predified classses or categories . in these problms , the model learns form a labeled dataset( where eac data point is associated with a class label ) and predicts the class label for new , unseen dat points .


## key charasteristics of a classficaion problem

- classes or labels : there are distict categorsi or labels that data points should be classfiied into . for exampe in a spam email classifier , the classes could be spam or not spam .

- descrete output : unlike regresson problems , which output continuous values , classficaion problems produce discrete outputs ( specific labels or categores ) .

- type of classficaion:
  - binary classficaon : only two classes.
  - multi-class classificaion : more than two classes
  - multi-label classficaion: each data point can belong to multiple classes at once

## set up loss funtion and optimzer

which loss funtion or optimizer shuld youuse ?

agian .. this is problem specific
for example for regression you might want MAE or MSE ( mean absolut error)
for classficion you might want binary cross entroy or cross entopy

binary cross entopy is a loss funtion used in machine leanring ,particularly in binary classfion broblmes . it measure the dissimilarity betweent he predicted probability distribution and actual binary labesl ( 0 or 1)

you may look at the formula online

as a reminder the loss funtion meaure how wrong your model predictin are
and for optizers , two of the most common and useful are SGD and Adam

* for the loss funtion we going to use
`torch.nn.BECWithLogistsLoss()`




```python
# setup the loss funtion
loss_fn = nn.BCEWithLogitsLoss()# has sigmoid activation funtion buit in
# the nn.BCEloss -> requres that the inputs have gone throught the sigmoid activation funtion



# setup the optimizer
optimizer = torch.optim.SGD(params = model_0.parameters(), lr = 0.1)
```

```python
# caculate accuracy - out of 100 example what percentage does our model get right?
def accuracy_fn(y_true, y_pred):
  correct = torch.eq(y_true, y_pred).sum().item()
  acc = (correct / len(y_pred)) * 100
  return acc
```

### building the training and test loop

```python
torch.manual_seed(42)
try:
  torch.cuda.manual_seed(42)
except:
  pass

epochs = 1000
# put dat to target device
X_train,y_train = X_train.to(device), y_train.to(device)
X_test, y_test = X_test.to(device), y_test.to(device)

# build training and evaluation loop
for epoch in range(epochs):
  model_0.train()
  # 1. forward pass
  y_logits = model_0(X_train).squeeze()
  y_pred = torch.round(torch.sigmoid(y_logits)) # turns logist -> pred probls -> pred lables
  # 2. cacluate loss and accuracy
  loss = loss_fn(y_logits, y_train)
  acc = accuracy_fn(y_true = y_train, y_pred = y_pred)
  # 3. optimizer zero grad
  optimizer.zero_grad()
  # do back propagation
  loss.backward()
  # optimizer step
  optimizer.step()
  ### testing
  model_0.eval()
  with torch.inference_mode():
    test_logits = model_0(X_test).squeeze()
    test_pred = torch.sigmoid(test_logits)
    test_pred = torch.round(test_pred)
    test_loss = loss_fn(test_logits, y_test)
    test_acc = accuracy_fn(y_true = y_test, y_pred = test_pred)
    #print out what happein
    if epoch % 10 == 0:
      print(f"epoch : {epoch} | loss : {loss:.5f} | acc : {acc:.2f}% | test_loss : {test_loss:.5f} | test_acc : {test_acc:.2f}")



```

### 6.2 building a model with non linearity
* linear = starght line
* non linear = non straight line


```python
# build a model with non - linear activation funtions
from torch import nn
class CirculeModelv2(nn.Module):
  def __init__(self):
    super().__init__()
    self.layer_1 = nn.Linear(in_features = 2, out_features = 10)
    self.layer_2 = nn.Linear(in_features = 10, out_features = 10)
    self.layer_3 = nn.Linear(in_features = 10, out_features = 1)
    self.relu = nn.ReLU()
  def forward(self, x):
    return self.layer_3(self.relu(self.layer_2(self.relu(self.layer_1(x)))))

model_3 = CirculeModelv2().to(device)
model_3
```

Artificial neural networks are a large combination of linear and non-linear funtions which are potentially able to find pattersn in data .

```python
# setup loss and optimzer
loss_fn = nn.BCEWithLogitsLoss()
optimzer = torch.optim.SGD(model_3.parameters(), lr = 0.1)
```

## trainig a model with non linearity


```python
# random seeds

torch.manual_seed(42)
try:
  torch.cuda.manual_seed(42)
except:
  pass
epochs = 10000

# put all data to the target device
X_train, y_train = X_train.to(device), y_train.to(device)
X_test, y_test = X_test.to(device), y_test.to(device)
# traingin loop
for epoch in range(epochs):
  model_3.train()
  y_logits = model_3(X_train).squeeze()
  y_pred = torch.round(torch.sigmoid(y_logits))
  loss = loss_fn(y_logits, y_train)
  acc = accuracy_fn(y_true = y_train, y_pred = y_pred)
  optimzer.zero_grad()
  loss.backward()
  optimzer.step()
### testing
model_3.eval()
with torch.inference_mode():
  test_pred = model_3(X_test).squeeze()
  test_loss = loss_fn(test_pred, y_test)
  print(f"epoch : {epoch} | loss : {loss:.5f} | test_loss : {test_loss:.5f}")
```
